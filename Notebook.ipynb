## .doc Invoices

```python
# Removing spaces from file names

import os

path =  '/Users/flatironschool/Final-Project/Invoices_WordDocs/'

filenames = os.listdir('/Users/flatironschool/Final-Project/Invoices_WordDocs/')

for filename in filenames:
    os.rename(os.path.join(path + filename), os.path.join(path + filename.replace(" ", "-")))
```

```python
# Converting Word docs to .txt docs

for file in os.listdir('/Users/flatironschool/Final-Project/Invoices_WordDocs'):

    os.system(f'/Applications/LibreOffice.app/Contents/MacOS/soffice --headless --convert-to txt:Text /Users/flatironschool/Final-Project/Invoices_WordDocs/{file} --outdir /Users/flatironschool/Final-Project/Invoices_WordDocs/output')
```

```python
# Putting imperfect invoices into folder called bad-files

import shutil

relevant_invoices = '/Users/flatironschool/Final-Project/Invoices_WordDocs/relevant-invoices'
bad_files = '/Users/flatironschool/Final-Project/Invoices_WordDocs/bad-files'

for file in os.listdir("/Users/flatironschool/Final-Project/Invoices_WordDocs/output"):
    with open(f'/Users/flatironschool/Final-Project/Invoices_WordDocs/output/{file}') as f:
        for line in f:
            if 'Artist:' in line:
                shutil.copy(f'/Users/flatironschool/Final-Project/Invoices_WordDocs/output/{file}', relevant_invoices)
            else:
                shutil.copy(f'/Users/flatironschool/Final-Project/Invoices_WordDocs/output/{file}', bad_files)             
                    
```

```python
# Figuring out unique items between bad-files and relevant-files

from filecmp import dircmp

dir_a = '/Users/flatironschool/Final-Project/Invoices_WordDocs/relevant-invoices'
dir_b = '/Users/flatironschool/Final-Project/Invoices_WordDocs/bad-files'

only_bads = dircmp(dir_a, dir_b)
```

```python
# Getting only the files unique to the bad-files folder

only_bads = only_bads.right_only
```

```python
# Putting the unique files from bad-files into a new folder called only_bad_files

only_bad_files = '/Users/flatironschool/Final-Project/Invoices_WordDocs/only-bad-files'


for file in only_bads:
    if file in os.listdir('/Users/flatironschool/Final-Project/Invoices_WordDocs/bad-files'):
        shutil.copy(f'/Users/flatironschool/Final-Project/Invoices_WordDocs/bad-files/{file}', only_bad_files)
        
```
